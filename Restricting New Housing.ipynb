```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# arbitrary data for supply curve, slope = 1
Quantity = np.array([0, 1, 2, 3, 3.5])
Price = np.array([0, 1, 2, 3, 3.5])

# convert to pandas DataFrame
d = {'Quantity': Quantity, 'Price': Price}
df_housing_supply = pd.DataFrame(d)
df_housing_supply
```

```python
# arbitrary data for supply curve shift
Quantity1 = np.array([0, 1, 2, 3])
Price1 = np.array([1, 2, 3, 4])

# convert to pandas DataFrame
d = {'Quantity': Quantity1, 'Price': Price1}
df_housing_supply_decrease = pd.DataFrame(d)

df_housing_supply_decrease
```

```python
# arbitrary data for demand curve, slope = -1
Quantity2 = np.array([3.5, 3, 2, 1, 0.5])
Price2 = np.array([0.5, 1, 2, 3, 3.5])

# convert to pandas DataFrame
d = {'Quantity': Quantity2, 'Price': Price2}
df_housing_demand = pd.DataFrame(d)

df_housing_demand
```

```python
# arbitrary data for demand curve increase
Quantity3 = np.array([4, 3, 2, 1])
Price3 = np.array([1, 2, 3, 4])

# convert to pandas DataFrame
d = {'Quantity': Quantity3, 'Price': Price3}
df_housing_demand_increase = pd.DataFrame(d)

df_housing_demand_increase
```

```python
# data for plumb lines of first equilibrium
dotted_one = pd.DataFrame(np.array([[0,2], [2,2], [2,0]]), 
                          columns=['xx', 'yy'])
dotted_one
```

```python
# data for plumb lines of second equilibrium
dotted_two = pd.DataFrame(np.array([[0,2.5], [2.5,2.5], [2.5,0]]), 
                          columns=['xx', 'yy'])
dotted_two
```

```python
# set figure size and style
fig = plt.figure(figsize=(8,6))
sns.set_style('darkgrid')

# plot axes
plt.axvline(ymin=0.07, ymax=0.93, color='black')
plt.axhline(xmin=0.07, xmax=0.93, color='black')

# plot supply and demand curves
plt.plot('Quantity', 'Price', data=df_housing_supply, label='Supply', 
         color='tomato', zorder=1, lw=2)

plt.plot('Quantity', 'Price', data=df_housing_demand, label='Demand', 
         color='lightseagreen', zorder=1, lw=2)

plt.plot('Quantity', 'Price', data=df_housing_demand_increase, 
         label='Demand increase', color='lightseagreen', linestyle='dashed',
         zorder=1, lw=2)

# plot equilibria and plumb lines
plt.scatter(x=2, y=2, zorder=2, color='grey', s=60, label='Old equilibrium')
plt.plot('xx', 'yy', data=dotted_one, linestyle='dotted', color='grey', label='', lw=3)

plt.scatter(x=2.5, y=2.5, zorder=2, color='royalblue', s=60, label='New equilibrium')
plt.plot('xx', 'yy', data=dotted_two, linestyle='dotted', color='royalblue', label='', lw=3)

# plot arrows
plt.arrow(3, 1, 0.3, 0.3, width=0.015, head_width=0.15, head_length=0.15, fc='k', ec='k')
plt.arrow(1, 3, 0.3, 0.3, width=0.015, head_width=0.15, head_length=0.15, fc='k', ec='k')

# label axes
plt.xlabel('Quantity')
plt.ylabel('Price')

# set axes limits
plt.xlim([-0.3, 4.3])
plt.ylim([-0.3, 4.3])

# title and legend
plt.title('St. Ives Housing Market', fontsize=16, pad=20)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.tight_layout()
fig.savefig('st-ives-housing-market.png', dpi=180);
```

```python
# data for plumb lines of third equilibrium
dotted_three = pd.DataFrame(np.array([[0,3], [2,3], [2,0]]), 
                          columns=['xx', 'yy'])
dotted_three
```

```python
# set figure size and style
fig = plt.figure(figsize=(12,8))
sns.set_style('darkgrid')

# ----- FIRST SUBPLOT ------
plt.subplot(1, 2, 1)

# plot axes
plt.axvline(ymin=0.07, ymax=0.93, color='black')
plt.axhline(xmin=0.07, xmax=0.93, color='black')

# plot supply and demand curves
plt.plot('Quantity', 'Price', data=df_housing_supply, label='Supply', 
         color='tomato', zorder=1, lw=2)

plt.plot('Quantity', 'Price', data=df_housing_demand, label='Demand', 
         color='lightseagreen', zorder=1, lw=2)

plt.plot('Quantity', 'Price', data=df_housing_demand_increase, 
         label='Demand increase', color='lightseagreen', linestyle='dashed',
         zorder=1, lw=2)

# plot equilibria and plumb lines
plt.scatter(x=2, y=2, zorder=2, color='grey', s=60, label='Old equilibrium')
plt.plot('xx', 'yy', data=dotted_one, linestyle='dotted', color='grey', label='', lw=3)

plt.scatter(x=2.5, y=2.5, zorder=2, color='royalblue', s=60, label='New equilibrium')
plt.plot('xx', 'yy', data=dotted_two, linestyle='dotted', color='royalblue', label='', lw=3)

# plot arrows
plt.arrow(3, 1, 0.3, 0.3, width=0.015, head_width=0.15, head_length=0.15, fc='k', ec='k')
plt.arrow(1, 3, 0.3, 0.3, width=0.015, head_width=0.15, head_length=0.15, fc='k', ec='k')

# axes labels
plt.xlabel('Quantity')
plt.ylabel('Price')

# axes limits
plt.xlim([-0.3, 4.3])
plt.ylim([-0.3, 4.3])

# title and legend
plt.title('Situation for Locals: Increase in Demand', fontsize=16, pad=20)
plt.legend(bbox_to_anchor=(0.5, -0.4), loc=8)
plt.tight_layout()

# ----- SECOND SUBPLOT -----
plt.subplot(1, 2, 2)

# plot axes
plt.axvline(ymin=0.07, ymax=0.93, color='black')
plt.axhline(xmin=0.07, xmax=0.93, color='black')

# plot supply and demand curves
plt.plot('Quantity', 'Price', data=df_housing_supply, label='Supply', 
         color='tomato', zorder=1, lw=2)

plt.plot('Quantity', 'Price', data=df_housing_supply_decrease, 
         label='Supply decrease', color='tomato', zorder=1, linestyle='dashed', lw=2)

plt.plot('Quantity', 'Price', data=df_housing_demand_increase, 
         label='Demand increase', color='lightseagreen', linestyle='dashed',
         zorder=1, lw=2)

# plot equilibria and plumb lines
plt.scatter(x=2, y=2, zorder=3, color='grey', s=60)

plt.scatter(x=2.5, y=2.5, zorder=2, color='grey', s=60, label='Old equilibria')
plt.plot('xx', 'yy', data=dotted_two, linestyle='dotted', color='grey', label='', lw=3)

plt.scatter(x=2, y=3, zorder=2, color='royalblue', s=60, label='New equilibrium')
plt.plot('xx', 'yy', data=dotted_three, linestyle='dotted', color='royalblue', label='', lw=3)

# plot arrows
plt.arrow(1, 1, -0.3, 0.3, width=0.015, head_width=0.15, head_length=0.15, fc='k', ec='k')
plt.arrow(3, 3, -0.3, 0.3, width=0.015, head_width=0.15, head_length=0.15, fc='k', ec='k')

# axes labels
plt.xlabel('Quantity')
plt.ylabel('Price')

# axes limits
plt.xlim([-0.3, 4.3])
plt.ylim([-0.3, 4.3])

# title and legend
plt.title('Response by Locals: Decrease in Supply', fontsize=16, pad=20)
plt.legend(bbox_to_anchor=(0.5, -0.4), loc=8)
plt.tight_layout()

# ----- End of second subplot -----

# figure title
plt.subplots_adjust(top=0.83)
fig.suptitle('St. Ives Housing Market: Pre-Legislation', fontsize=20)
fig.savefig('st-ives-housing-market-pre-legislation.png', dpi=180);
```

```python
# set figure size and style
fig = plt.figure(figsize=(12,8))
sns.set_style('darkgrid')

# ----- FIRST SUBPLOT -----
plt.subplot(1, 2, 1)

# plot axes
plt.axvline(ymin=0.07, ymax=0.93, color='black')
plt.axhline(xmin=0.07, xmax=0.93, color='black')

# plot supply and demand curves
plt.plot('Quantity', 'Price', data=df_housing_supply, label='Supply', 
         color='tomato', zorder=1, lw=2)

plt.plot('Quantity', 'Price', data=df_housing_supply_decrease, 
         label='Supply decrease', color='tomato', zorder=1, linestyle='dashed', lw=2)

plt.plot('Quantity', 'Price', data=df_housing_demand_increase, label='Demand', color='lightseagreen',
         linestyle='dashed', zorder=1, lw=2)

# plot equilibria and plumb lines
plt.scatter(x=2, y=2, zorder=3, color='grey', s=60)

plt.scatter(x=2.5, y=2.5, zorder=2, color='grey', s=60, label='Old equilibria')
plt.plot('xx', 'yy', data=dotted_two, linestyle='dotted', color='grey', label='', lw=3)

plt.scatter(x=2, y=3, zorder=2, color='royalblue', s=60, label='New equilibrium')
plt.plot('xx', 'yy', data=dotted_three, linestyle='dotted', color='royalblue', label='', lw=3)

# plot arrows
plt.arrow(1, 1, -0.3, 0.3, width=0.015, head_width=0.15, head_length=0.15, fc='k', ec='k')
plt.arrow(3, 3, -0.3, 0.3, width=0.015, head_width=0.15, head_length=0.15, fc='k', ec='k')

# axes labels
plt.xlabel('Quantity')
plt.ylabel('Price')

# axes limits
plt.xlim([-0.3, 4.3])
plt.ylim([-0.3, 4.3])

# title and legend
plt.title('Situation for Non-residents: Decrease in Supply', fontsize=16, pad=20)
plt.legend(bbox_to_anchor=(0.5, -0.4), loc=8)
plt.tight_layout()

# ----- SECOND SUBPLOT -----
plt.subplot(1, 2, 2)

# plot axes
plt.axvline(ymin=0.07, ymax=0.93, color='black')
plt.axhline(xmin=0.07, xmax=0.93, color='black')

# plot supply and demand curves
plt.plot('Quantity', 'Price', data=df_housing_supply_decrease, 
         label='Supply decrease', color='tomato', zorder=1, linestyle='dashed', lw=2)

plt.plot('Quantity', 'Price', data=df_housing_demand_increase,
         label='Demand increase', color='lightseagreen', linestyle='dashed',
         zorder=1, lw=2)
plt.plot('Quantity', 'Price', data=df_housing_demand_decrease,
         label='Demand decrease', color='lightseagreen', zorder=1,
         linestyle='dashdot', lw=2)

# plot equilibria and plumb lines
plt.scatter(x=2, y=2, zorder=3, color='grey', s=60)
plt.scatter(x=2.5, y=2.5, zorder=2, color='grey', s=60)

plt.scatter(x=2, y=3, zorder=2, color='grey', s=60, label='Old equilibria')
plt.plot('xx', 'yy', data=dotted_three, linestyle='dotted', color='grey', label='', lw=3)

plt.scatter(x=1.75, y=2.75, zorder=2, color='royalblue', s=60, label='New equilibrium')
plt.plot('xx', 'yy', data=dotted_four, linestyle='dotted', color='royalblue', label='', lw=3)

# plot arrows
plt.arrow(1.5, 3.5, -0.10, -0.10, width=0.015, head_width=0.12, head_length=0.12, fc='k', ec='k')
plt.arrow(3.5, 1.5, -0.10, -0.10, width=0.015, head_width=0.12, head_length=0.12, fc='k', ec='k')

# axes labels
plt.xlabel('Quantity')
plt.ylabel('Price')

# axes limits
plt.xlim([-0.3, 4.3])
plt.ylim([-0.3, 4.3])

# title and legend
plt.title('Response by All: Decrease in Demand', fontsize=16, pad=20)
plt.legend(bbox_to_anchor=(0.5, -0.4), loc=8)
plt.tight_layout()

# ----- End of second subplot -----

# figure title
plt.subplots_adjust(top=0.83)
fig.suptitle('St. Ives Housing Market: Post-Legislation', fontsize=20)
fig.savefig('st-ives-housing-market-post-legislation.png', dpi=180);
```

```python
# real data on St. Ives, from Financial Times
# numbers rounded to nearest thousand British pound
# https://www.ft.com/content/6abb85e8-c349-11e9-ae6e-a26d1d0455f4
df_dots_2016 = pd.DataFrame(np.array([[310,0], [310,323], [0,323]]), 
                          columns=['xx', 'yy'])

df_dots_2018 = pd.DataFrame(np.array([[270,0], [270,352], [0,352]]), 
                          columns=['xx', 'yy'])
```

```python
# set figure size and style
fig = plt.figure(figsize=(12,8))
sns.set_style('darkgrid')

# ----- FIRST SUBPLOT ------
plt.subplot(1, 2, 1)

# plot axes
plt.axvline(ymin=0.07, ymax=0.93, color='black')
plt.axhline(xmin=0.07, xmax=0.93, color='black')

# plot supply and demand curves
plt.plot('Quantity', 'Price', data=df_housing_supply, color='tomato',
         zorder=1, lw=2, label='')

plt.plot('Quantity', 'Price', data=df_housing_demand_increase, 
         color='lightseagreen', linestyle='dashed', zorder=1, lw=2, label='')

plt.plot('Quantity', 'Price', data=df_housing_supply_decrease, color='tomato',
         zorder=1, linestyle='dashed', lw=2, label='')

plt.plot('Quantity', 'Price', data=df_housing_demand_decrease,
         color='lightseagreen', zorder=1, linestyle='dashdot', lw=2, label='')


# plot equilibria and plumb lines
plt.scatter(x=2.5, y=2.5, zorder=2, color='grey', s=60, 
            label='Pre-legislation equilibrium')
plt.plot('xx', 'yy', data=dotted_two, linestyle='dotted', color='grey', 
         label='', lw=3)

plt.scatter(x=1.75, y=2.75, zorder=2, color='royalblue', s=60, 
            label='Post-legislation equilibrium')
plt.plot('xx', 'yy', data=dotted_four, linestyle='dotted', 
         color='royalblue', label='', lw=3)

# axes labels
plt.xlabel('Quantity')
plt.ylabel('Price')

# axes limits
plt.xlim([-0.3, 4.3])
plt.ylim([-0.3, 4.3])

# title and legend
plt.title('Theory', fontsize=16, pad=20)
plt.legend(bbox_to_anchor=(0.5, -0.25), loc=8)
plt.tight_layout()

# ----- SECOND SUBPLOT -----
plt.subplot(1, 2, 2)

# plot axes
plt.axvline(ymin=0.07, ymax=0.93, color='black')
plt.axhline(xmin=0.07, xmax=0.93, color='black')

# plot equilibria and plumb lines
plt.scatter(x=310, y=323, zorder=2, color='grey', s=60, label='2016 equilibrium, pre-legislation')
plt.plot('xx', 'yy', data=df_dots_2016, linestyle='dotted', color='grey', label='', lw=3)

plt.scatter(x=270, y=352, zorder=2, color='royalblue', s=60, label='2018 equilibrium, post-legislation')
plt.plot('xx', 'yy', data=df_dots_2018, linestyle='dotted', color='royalblue', label='', lw=3)

# axes labels
plt.xlabel('Quantity')
plt.ylabel('Price in thousands of pounds')

# axes limits
plt.xlim([-38, 500])
plt.ylim([-38, 500])

# title and legend
plt.title('Reality', fontsize=16, pad=20)
plt.legend(bbox_to_anchor=(0.5, -0.25), loc=8)
plt.tight_layout()

# ----- End of second subplot -----

# figure title
plt.subplots_adjust(top=0.83)
fig.suptitle('St. Ives Housing Market: Before & After Legislation', fontsize=20)
fig.savefig('st-ives-housing-market-theoretical-and-real.png', dpi=180);
```
